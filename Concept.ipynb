# Understanding `matplotlib.pyplot`

## What is `pyplot`?

`pyplot` is a module (sub-library) of Matplotlib that provides a simple and interactive way to create, customize, and display graphs.

It contains ready-to-use functions for creating different types of charts without writing complex code.

Almost every Matplotlib program begins with:

```python
import matplotlib.pyplot as plt
```

Here:

- `matplotlib` → The complete visualization library.
- `pyplot` → A module inside Matplotlib used for plotting.
- `plt` → A short alias for `matplotlib.pyplot`.

---

## Why Do We Need `pyplot`?

Imagine you want to draw a graph.

Without `pyplot`, you would need to manually create a figure, add axes, configure the plotting area, and then draw the graph. This involves multiple steps and more code.

`pyplot` simplifies this process by providing easy-to-use functions that perform these tasks for you.

Instead of writing many lines of code, you can create a graph in just a few lines.

For example:

```python
plt.plot(x, y)
plt.show()
```

This is one of the main reasons why beginners and data analysts commonly use `pyplot`.

---

## What Can We Do Using `pyplot`?

`pyplot` provides functions to create many different types of visualizations, including:

- Line Charts
- Bar Charts
- Scatter Plots
- Histograms
- Pie Charts
- Box Plots
- Subplots

It also allows us to customize graphs by adding:

- Titles
- Axis labels
- Legends
- Grid lines
- Colors
- Markers
- Figure size
- Line styles

## Why Do We Write `import matplotlib.pyplot as plt`?

We use:

```python
import matplotlib.pyplot as plt
```

instead of

```python
import matplotlib.pyplot
```

because:

- It makes the code shorter and easier to read.
- `plt` is the standard alias used by the Python community.
- Almost every Matplotlib tutorial, documentation page, and project uses `plt`, making code more consistent and easier to understand.

---

## Key Points

- `pyplot` is a module inside the Matplotlib library.
- It provides simple functions to create and customize graphs.
- It automatically handles many plotting tasks behind the scenes.
- It is the most commonly used interface for creating visualizations in Python.
- The standard import statement is:

```python
import matplotlib.pyplot as plt
```

Many beginners think pyplot draws the graph. More accurately, pyplot is an interface that lets you create and manage figures, axes, and plots. It provides convenient functions (like plot(), bar(), and show()) that internally work with Matplotlib's figure and axes objects.

# Commonly Used `pyplot` Functions

The `pyplot` module provides many built-in functions to create and customize graphs. These functions help us display data clearly and make graphs more informative.

Below are some of the most frequently used functions.

---

## 1. `plt.plot()`

**Purpose:** Creates a line chart by connecting data points with straight lines.

### Syntax

```python
plt.plot(x, y)
```

### Example

```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
```

**When to use:**

- Sales over time
- Stock prices
- Temperature changes
- Population growth
- Any continuous data

---

## 2. `plt.bar()`

**Purpose:** Creates a vertical bar chart.

### Syntax

```python
plt.bar(x, height)
```

### Example

```python
products = ["A", "B", "C"]
sales = [120, 150, 100]

plt.bar(products, sales)
```

**When to use:**

- Compare categories
- Compare product sales
- Student marks
- Company profits

---

## 3. `plt.scatter()`

**Purpose:** Creates a scatter plot where each point is displayed separately.

### Syntax

```python
plt.scatter(x, y)
```

### Example

```python
height = [150, 160, 170, 180]
weight = [45, 55, 65, 75]

plt.scatter(height, weight)
```

**When to use:**

- Find relationships between two variables
- Detect clusters
- Detect outliers

---

## 4. `plt.hist()`

**Purpose:** Creates a histogram to show the distribution of numerical data.

### Syntax

```python
plt.hist(data)
```

### Example

```python
marks = [45, 50, 60, 65, 70, 70, 75, 80, 85]

plt.hist(marks)
```

**When to use:**

- Age distribution
- Salary distribution
- Exam score distribution

---

## 5. `plt.pie()`

**Purpose:** Creates a pie chart showing proportions of different categories.

### Syntax

```python
plt.pie(values)
```

### Example

```python
expenses = [40, 30, 20, 10]

plt.pie(expenses)
```

**When to use:**

- Budget allocation
- Market share
- Expense breakdown

---

## 6. `plt.xlabel()`

**Purpose:** Adds a label to the x-axis.

### Syntax

```python
plt.xlabel("Month")
```

---

## 7. `plt.ylabel()`

**Purpose:** Adds a label to the y-axis.

### Syntax

```python
plt.ylabel("Sales")
```

---

## 8. `plt.title()`

**Purpose:** Adds a title to the graph.

### Syntax

```python
plt.title("Monthly Sales")
```

---

## 9. `plt.legend()`

**Purpose:** Displays a legend to identify different datasets plotted on the same graph.

### Syntax

```python
plt.legend()
```

**When to use:**

When plotting two or more datasets on the same graph.

---

## 10. `plt.grid()`

**Purpose:** Displays grid lines on the graph.

### Syntax

```python
plt.grid(True)
```

**Why use it?**

Grid lines make it easier to estimate values and compare points on the graph.

---

## 11. `plt.show()`

**Purpose:** Displays the completed graph.

### Syntax

```python
plt.show()
```

Without `plt.show()`, a Python script may not display the graph. In Jupyter Notebook, plots are often displayed automatically, but using `plt.show()` is still considered good practice.
