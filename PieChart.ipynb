# Pie Chart

## What is a Pie Chart?

A **Pie Chart** is a circular graph divided into slices.

Each slice represents a category's contribution to the whole.

The complete circle always represents **100%** of the data.

The size of each slice is proportional to its value.

---

## Why Do We Need a Pie Chart?

Suppose a student spends time every day on different activities.

| Activity | Hours |
|----------|-------|
| Study | 6 |
| Sleep | 8 |
| Exercise | 2 |
| Entertainment | 3 |
| Others | 5 |

Although the table contains all the information, it is difficult to understand which activity occupies the largest portion of the day.

A Pie Chart immediately shows:

- Largest category
- Smallest category
- Percentage contribution
- Overall distribution

---

## When Should We Use a Pie Chart?

Use a Pie Chart when:

- You want to show proportions.
- Categories represent parts of one whole.
- The total has meaningful significance.
- There are a small number of categories (generally 3–6).

---

## Real-Life Examples

- Budget Distribution
- Monthly Expenses
- Market Share
- Election Votes
- Browser Usage
- Department-wise Budget

---

## When NOT to Use a Pie Chart?

Avoid Pie Charts when:

- There are too many categories.
- Exact comparisons are required.
- Categories do not represent a complete whole.

In such cases, use a **Bar Chart** instead.

---

## Syntax

```python
plt.pie(x)
```

Where

- `x` → Numerical values

Common parameters:

- labels
- colors
- autopct
- startangle
- explode
- shadow

---

## Advantages

- Easy to understand percentages.
- Good for showing proportions.
- Attractive visualization.

---

## Limitations

- Difficult to compare many categories.
- Not suitable for large datasets.
- Small slices become difficult to read.

````python
import matplotlib.pyplot as plt

expenses = [25000, 12000, 8000, 5000]

categories = ["Rent", "Food", "Travel", "Shopping"]

plt.pie(expenses, labels=categories)

plt.show()
````

# Professional Pie Chart

````python
import matplotlib.pyplot as plt

expenses = [25000, 12000, 8000, 5000]

categories = [
    "Rent",
    "Food",
    "Travel",
    "Shopping"
]

colors = [
    "royalblue",
    "orange",
    "green",
    "crimson"
]

explode = (
    0.1,
    0,
    0,
    0
)

plt.figure(figsize=(8,8))

plt.pie(

    expenses,

    labels=categories,

    colors=colors,

    explode=explode,

    autopct="%1.1f%%",

    startangle=90,

    shadow=True,

    wedgeprops={
        "edgecolor":"black",
        "linewidth":2
    }

)

plt.title(
    "Monthly Expense Distribution",
    fontsize=16,
    fontweight="bold"
)

plt.legend(
    title="Expense Categories",
    loc="upper right"
)

plt.show()
````

| Property   | Purpose                  |
| ---------- | ------------------------ |
| labels     | Displays category names  |
| colors     | Slice colors             |
| explode    | Separates slice from pie |
| autopct    | Shows percentage         |
| startangle | Rotates chart            |
| shadow     | Adds shadow              |
| wedgeprops | Border customization     |
| figsize    | Figure size              |
| title()    | Graph title              |
| legend()   | Displays legend          |


# Donut Chart

````python
import matplotlib.pyplot as plt

expenses = [25,15,20,40]

labels = [
    "Food",
    "Travel",
    "Shopping",
    "Rent"
]

plt.figure(figsize=(7,7))

plt.pie(
    expenses,
    labels=labels,
    autopct="%1.1f%%",
    wedgeprops={
        "width":0.45,
        "edgecolor":"white"
    }
)

plt.title("Expense Distribution")

plt.show()
````

| Pie Chart               | Bar Chart                 |
| ----------------------- | ------------------------- |
| Shows percentages       | Compares values           |
| Circular                | Rectangular bars          |
| Parts of whole          | Independent categories    |
| Best for 3–6 categories | Works for many categories |


| Pie Chart         | Histogram                 |
| ----------------- | ------------------------- |
| Shows proportions | Shows distributions       |
| Categorical parts | Continuous numerical data |
| Circular          | Rectangular bins          |
