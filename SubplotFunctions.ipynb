# Subplots in Matplotlib

## What is a Subplot?

A **subplot** is a smaller plot that exists inside a larger Figure.

Instead of creating multiple separate figures, Matplotlib allows us to place multiple plots inside the same Figure.

Each subplot has its own:

- X-axis
- Y-axis
- Title
- Grid
- Labels
- Data

---

## Why Do We Need Subplots?

Suppose you are analyzing a company's sales.

You want to display:

- Monthly Sales
- Monthly Profit
- Monthly Expenses
- Monthly Revenue

Creating four separate figures makes comparison difficult.

Instead, we place all four charts inside one Figure using **subplots**.

This makes comparison easier and saves space.

---

## Real-Life Applications

- Sales Dashboard
- Stock Market Analysis
- Medical Reports
- Machine Learning Visualizations
- Exploratory Data Analysis (EDA)
- Business Reports

---

## Types of Subplots

Matplotlib provides two main ways:

1. `plt.subplot()`
2. `plt.subplots()`

Although both create subplots, `plt.subplots()` is the modern and recommended approach.

---

## Hierarchy

Figure

# plt.subplot()

`plt.subplot()` creates one subplot at a time.

Syntax:

```python
plt.subplot(rows, columns, index)
```

Where:

- rows → Number of rows
- columns → Number of columns
- index → Position of current subplot

The index starts from **1**, not 0.

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

# First subplot
plt.subplot(1,2,1)

plt.plot([1,2,3],[2,5,8])

plt.title("Sales")

# Second subplot
plt.subplot(1,2,2)

plt.bar(
    ["A","B","C"],
    [20,35,15]
)

plt.title("Products")

plt.show()
````

````python
#plt.subplot(nrows,ncols, index)

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1) #1rows,2 columns, 1st subplot
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1, 2, 2) #1row, 2column , 2nd subplot
plt.bar(x,y)
plt.title('Bar chart')

plt.tight_layout()
plt.show()
````

# plt.subplots()

`plt.subplots()` creates the Figure and all Axes at the same time.

It follows the Object-Oriented (OO) approach.

This is the preferred method in professional projects because it is cleaner, easier to read, and easier to manage.

# Single Plot

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(
    [1,2,3],
    [2,5,8]
)

plt.show()
````

# Two Subplots

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(
    1,
    2,
    figsize=(10,5)
)

ax[0].plot(
    [1,2,3],
    [2,5,8]
)

ax[0].set_title("Sales")

ax[1].bar(
    ["A","B","C"],
    [20,35,15]
)

ax[1].set_title("Products")

plt.show()
````

# 2 x 2 grid

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(
    2,
    2,
    figsize=(10,8)
)

# Plot 1
ax[0,0].plot([1,2,3],[2,4,6])
ax[0,0].set_title("Line")

# Plot 2
ax[0,1].bar(
    ["A","B","C"],
    [10,20,15]
)
ax[0,1].set_title("Bar")

# Plot 3
ax[1,0].scatter(
    [1,2,3],
    [3,5,2]
)
ax[1,0].set_title("Scatter")

# Plot 4
ax[1,1].hist(
    [2,3,4,5,6,6,7,8]
)
ax[1,1].set_title("Histogram")

plt.tight_layout()

plt.show()
````

````python
import matplotlib.pyplot as plt

#fig , ax = plt.subplots(nrows, ncols, figsize = (width, height))
fig, ax = plt.subplots(1,2, figsize = (10,5) )

x = [1,2,3,4]
y = [10,20,15,25]



ax[0].plot(x,y,color = 'red')
ax[0].set_title('Line Plot')

ax[1].bar(x,y, color = 'green')
ax[1].set_title('Bar Chart')

plt.savefig("SubplotFunction.png", dpi=300)
plt.tight_layout()
plt.show()

````

| Function             | Purpose                         |
| -------------------- | ------------------------------- |
| `plt.subplot()`      | Create one subplot at a time    |
| `plt.subplots()`     | Create Figure and Axes together |
| `plt.tight_layout()` | Automatically adjust spacing    |
| `figsize`            | Set Figure size                 |
| `ax.plot()`          | Line chart                      |
| `ax.bar()`           | Bar chart                       |
| `ax.scatter()`       | Scatter plot                    |
| `ax.hist()`          | Histogram                       |
| `ax.set_title()`     | Title                           |
| `ax.set_xlabel()`    | X-axis label                    |
| `ax.set_ylabel()`    | Y-axis label                    |
| `ax.grid()`          | Display grid                    |
| `ax.legend()`        | Display legend                  |


| `plt.subplot()`               | `plt.subplots()`                          |
| ----------------------------- | ----------------------------------------- |
| Creates one subplot at a time | Creates all subplots together             |
| Uses pyplot interface         | Uses Object-Oriented interface            |
| Older style                   | Modern and recommended                    |
| Better for simple layouts     | Better for dashboards and complex figures |


# Interview Questions
# Q1. What is a subplot?

A subplot is an individual plot (Axes) inside a larger Figure. It allows multiple charts to be displayed in one window.

# Q2. Difference between subplot() and subplots()?
subplot() creates one subplot at a time using the pyplot interface.
subplots() creates the Figure and all Axes together using the Object-Oriented interface.

# Q3. What does plt.tight_layout() do?

It automatically adjusts the spacing between subplots so that titles, labels, and tick marks do not overlap.

# Q4. Why is plt.subplots() preferred?

Because it is cleaner, follows the Object-Oriented approach, and is easier to manage when working with multiple plots.
