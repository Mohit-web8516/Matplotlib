# Histogram

## What is a Histogram?

A **Histogram** is a graph used to display the **distribution of continuous numerical data**.

Instead of showing individual values, it groups similar values into ranges called **bins**.

The height of each bar represents the **frequency** (count) of values that fall within that range.

Unlike a Bar Chart, the bars in a Histogram **touch each other** because the data is continuous.

---

## Why Do We Need a Histogram?

Suppose a teacher has marks of 100 students.

| Marks |
|-------|
| 45 |
| 52 |
| 65 |
| 68 |
| 70 |
| 75 |
| ... |

Reading all the marks individually does not tell us:

- Which range has the most students?
- Are most students scoring high or low?
- Is the data spread evenly?
- Are there any unusual values?

A Histogram answers these questions by grouping the marks into ranges.

---

## What is a Bin?

A **Bin** is an interval (range) of values.

For example,

If marks range from **0 to 100**, bins may be:

0–20

20–40

40–60

60–80

80–100

The Histogram counts how many values fall into each bin.

---

## When Should We Use a Histogram?

Use a Histogram when:

- Data is continuous.
- You want to study the distribution of data.
- You want to identify the most common range.
- You want to detect skewness.
- You want to detect outliers.

---

## Real-Life Examples

- Student Marks
- Employee Salaries
- Customer Ages
- Heights
- Weights
- House Prices
- Daily Temperatures

---

## When NOT to Use a Histogram?

Avoid Histograms when:

- Data is categorical.
- You want to compare categories.
- You want to show percentages.

For categorical data, use a **Bar Chart** instead.

---

## Syntax

```python
plt.hist(data)
```

Common parameters:

- bins
- color
- edgecolor
- linewidth
- alpha

---

## Advantages

- Shows data distribution.
- Detects skewness.
- Detects outliers.
- Shows concentration of values.
- Easy to understand.

---

## Limitations

- Cannot compare categories.
- Choice of bins affects the appearance.
- Does not show relationships between variables.

````python
import matplotlib.pyplot as plt

marks = [
    45,50,52,55,58,
    60,62,65,67,70,
    72,75,78,80,82,
    85,88,90,92,95
]

plt.hist(marks)

plt.show()
````

# Professional Histogram

````python
import matplotlib.pyplot as plt

marks = [
    45,50,52,55,58,
    60,62,65,67,70,
    72,75,78,80,82,
    85,88,90,92,95
]

plt.figure(figsize=(8,5))

plt.hist(

    marks,

    bins=5,

    color="royalblue",

    edgecolor="black",

    linewidth=2,

    alpha=0.8,

    label="Students"

)

plt.title(
    "Distribution of Student Marks",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Marks",
    fontsize=12
)

plt.ylabel(
    "Frequency",
    fontsize=12
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.5
)

plt.legend()

plt.show()
````

# Histogram with Custom Bins

````python
import matplotlib.pyplot as plt

marks = [
    40,42,45,48,50,
    55,58,60,63,65,
    68,70,72,75,78,
    80,82,85,88,90
]

bins = [40,50,60,70,80,90]

plt.figure(figsize=(8,5))

plt.hist(

    marks,

    bins=bins,

    color="green",

    edgecolor="black"

)

plt.title("Student Marks Distribution")

plt.xlabel("Marks")

plt.ylabel("Frequency")

plt.show()
````

| Property  | Purpose                      |
| --------- | ---------------------------- |
| bins      | Number or range of intervals |
| color     | Bar color                    |
| edgecolor | Border color                 |
| linewidth | Border thickness             |
| alpha     | Transparency                 |
| figsize   | Figure size                  |
| title()   | Graph title                  |
| xlabel()  | X-axis label                 |
| ylabel()  | Y-axis label                 |
| legend()  | Displays legend              |
| grid()    | Shows grid                   |


| Histogram          | Bar Chart           |
| ------------------ | ------------------- |
| Continuous Data    | Categorical Data    |
| Bars touch         | Bars have gaps      |
| Shows distribution | Compares categories |
| Uses bins          | Uses categories     |


| Histogram    | Line Chart      |
| ------------ | --------------- |
| Distribution | Trend           |
| Frequency    | Ordered values  |
| Bars         | Connected lines |


| Histogram    | Scatter Plot            |
| ------------ | ----------------------- |
| One variable | Two variables           |
| Distribution | Relationship            |
| Frequency    | Individual observations |
