# Bar Chart

## What is a Bar Chart?

A **Bar Chart** is a graph used to compare values across different categories.

Each category is represented by a rectangular bar, and the height (or length) of the bar represents its value.

Unlike a Line Chart, the bars are **not connected** because every category is independent.

---

## Why Do We Need a Bar Chart?

Suppose a company sells four different products.

| Product | Sales |
|---------|------:|
| Laptop | 120 |
| Phone | 180 |
| Tablet | 90 |
| Watch | 150 |

Although we can read the table, it is much easier to compare sales using a Bar Chart.

A Bar Chart immediately tells us:

- Which product sold the most.
- Which product sold the least.
- The difference in sales between products.

---

## When Should We Use a Bar Chart?

Use a Bar Chart when:

- Data is categorical.
- Categories are independent.
- You want to compare values.

---

## Real-Life Examples

- Product Sales
- Student Marks
- Population of Cities
- Company Revenue
- Olympic Medals
- Number of Employees

---

## When NOT to Use a Bar Chart?

Avoid Bar Charts when:

- Showing trends over time.
- Showing distribution of continuous data.
- Showing relationships between numerical variables.

---

## Syntax

```python
plt.bar(x, height)
```

Where:

- `x` → Categories
- `height` → Values

Each category is represented by one bar.

---

## Advantages

- Easy to compare categories.
- Easy to understand.
- Suitable for categorical data.
- Works well with small and medium datasets.

---

## Limitations

- Not suitable for continuous data.
- Cannot show relationships between variables.
- Too many bars can make the chart difficult to read.

````python
import matplotlib.pyplot as plt

products = ["Laptop", "Phone", "Tablet", "Watch"]

sales = [120, 180, 90, 140]

plt.bar(products, sales,color = "Pink")

plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Units Sold")

plt.show()
````

# Professional Bar Chart

````python
import matplotlib.pyplot as plt

products = ["Laptop", "Phone", "Tablet", "Watch"]

sales = [120, 180, 90, 150]

plt.figure(figsize=(8,5))

plt.bar(
    products,
    sales,

    color=["royalblue","orange","green","crimson"],

    edgecolor="black",

    linewidth=2,

    width=0.6,

    alpha=0.85,

    label="2026 Sales"
)

plt.title(
    "Product Sales Analysis",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel("Products", fontsize=12)

plt.ylabel("Units Sold", fontsize=12)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.5
)

plt.legend()

plt.ylim(0,220)

plt.show()
````

| Property    | Purpose              |
| ----------- | -------------------- |
| `color`     | Changes bar color    |
| `edgecolor` | Border color of bars |
| `linewidth` | Thickness of border  |
| `width`     | Width of bars        |
| `alpha`     | Transparency         |
| `label`     | Legend text          |
| `figsize`   | Figure size          |
| `title()`   | Adds graph title     |
| `xlabel()`  | Adds x-axis label    |
| `ylabel()`  | Adds y-axis label    |
| `grid()`    | Displays grid lines  |
| `ylim()`    | Sets y-axis range    |
| `legend()`  | Displays legend      |


# Horizontal Bar Chart

````python
import matplotlib.pyplot as plt

products = ["Laptop", "Phone", "Tablet", "Watch"]

sales = [120,180,90,150]

plt.figure(figsize=(8,5))

plt.barh(
    products,
    sales,
    color="teal",
    edgecolor="black"
)

plt.title("Horizontal Bar Chart")

plt.xlabel("Sales")

plt.ylabel("Products")

plt.grid(axis="x", linestyle="--")

plt.show()
````

# Multiple Bar Chart (Grouped Bar Chart)

````python
import matplotlib.pyplot as plt
import numpy as np

products = ["Laptop","Phone","Tablet","Watch"]

sales_2025 = [100,150,80,120]

sales_2026 = [120,180,90,150]

x = np.arange(len(products))

width = 0.35

plt.figure(figsize=(9,5))

plt.bar(
    x - width/2,
    sales_2025,
    width=width,
    label="2025"
)

plt.bar(
    x + width/2,
    sales_2026,
    width=width,
    label="2026"
)

plt.xticks(x, products)

plt.title("Sales Comparison")

plt.xlabel("Products")

plt.ylabel("Sales")

plt.legend()

plt.grid(axis="y", linestyle="--")

plt.show()
````

# Stacked Bar Chart

````python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr"]

online = [40,60,55,70]

offline = [30,35,40,45]

plt.figure(figsize=(8,5))

plt.bar(
    months,
    online,
    label="Online"
)

plt.bar(
    months,
    offline,
    bottom=online,
    label="Offline"
)

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales")

plt.legend()

plt.show()
````

| Bar Chart              | Line Chart           |
| ---------------------- | -------------------- |
| Compares categories    | Shows trends         |
| Independent categories | Ordered data         |
| Uses bars              | Uses connected lines |


| Bar Chart          | Histogram             |
| ------------------ | --------------------- |
| Categorical Data   | Continuous Data       |
| Bars have gaps     | Bars touch each other |
| Compare categories | Show distribution     |


# Interview Questions
Q1. When should you use a Bar Chart?

To compare values across independent categories.

Q2. What is the difference between a Bar Chart and a Histogram?

A Bar Chart compares categorical data, while a Histogram shows the distribution of continuous numerical data.

Q3. What is the difference between plt.bar() and plt.barh()?
plt.bar() creates vertical bars.
plt.barh() creates horizontal bars.
Q4. When should you use a Horizontal Bar Chart?

When category names are long or when a horizontal layout is easier to read.
