# Matplotlib - Complete Beginner Guide

## What is Matplotlib?

Matplotlib is one of the most popular Python libraries used for **data visualization**. It allows us to convert numerical data into graphs and charts, making it easier to understand patterns, trends, and relationships.

Instead of looking at hundreds of rows of numbers, we can visualize the data in just a few seconds.

Examples of graphs created using Matplotlib:

- Line Chart
- Bar Chart
- Scatter Plot
- Histogram
- Pie Chart
- Box Plot
- Subplots

Matplotlib is widely used in:

- Data Analytics
- Data Science
- Machine Learning
- Research
- Business Intelligence

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Mon","Tue","Wed","Thur","Fri"] #x-axis
y = [10,15,9,28,12] #y-axis

plt.plot(x,y)
plt.title("Bakery Sales This Week")

plt.xlabel('Week Days')
plt.ylabel("Sales Per Day")

plt.show()
```

```python
#plt.plot(x,y, color='color name',linestyle = 'line_style',linewidth=value, marker = 'marker symbol',label = 'label name')
#plt.xlabel('text')
import matplotlib.pyplot as plt 

months = [1,2,3,4]
sales = [1000,1500,1200,1800]
plt.plot(months,sales,color='blue',linestyle='--',linewidth = 2, marker = 'o', label = '2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Data Report')
plt.legend(loc='upper left', fontsize = 12)
plt.grid(color= 'gray', linestyle = ':',linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ["Laptop", "Phone", "Tablet", "Watch"]

sales = [120, 180, 90, 140]

plt.bar(products, sales,color = "red")

plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Units Sold")

plt.show()
```

```python
import matplotlib.pyplot as plt

products = ["Laptop", "Phone", "Tablet", "Watch"]
sales = [120, 180, 90, 140]

plt.pie(
    sales,
    labels=products,
    autopct="%1.1f%%"
)

plt.title("Product Sales Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [40, 48, 55, 63, 70, 78, 85, 95]

# Create Figure
plt.figure(figsize=(8, 5))

# Scatter Plot
plt.scatter(
    study_hours,
    marks,

    # Point color
    color="red",

    # Point size
    s=150,

    # Marker style
    marker="^",

    # Transparency
    alpha=0.8,

    # Border color
    edgecolor="black",

    # Border width
    linewidth=1.5,

    # Legend label
    label="Student Marks"
)

# Title
plt.title(
    "Study Hours vs Marks",
    fontsize=16,
    color="blue",
    pad=15
)

# X-axis Label
plt.xlabel(
    "Study Hours",
    fontsize=12
)

# Y-axis Label
plt.ylabel(
    "Marks",
    fontsize=12
)

# Grid
plt.grid(
    True,
    linestyle="--",
    alpha=0.6
)

# Legend
plt.legend(
    loc="upper left",
    fontsize=10
)

# X-axis Limit
plt.xlim(0, 9)

# Y-axis Limit
plt.ylim(30, 100)

# Show Plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dataset 1
study_hours_class_A = [1, 2, 3, 4, 5, 6, 7]
marks_class_A = [42, 50, 58, 66, 73, 82, 90]

# Dataset 2
study_hours_class_B = [1, 2, 5, 4, 5, 6, 7]
marks_class_B = [35, 45, 52, 60, 67, 75, 83]

# Create Figure
plt.figure(figsize=(8, 6))

# Scatter Plot for Class A
plt.scatter(
    study_hours_class_A,
    marks_class_A,
    color="blue",
    marker="o",
    s=120,
    alpha=0.8,
    edgecolor="black",
    linewidth=1.5,
    label="Class A"
)

# Scatter Plot for Class B
plt.scatter(
    study_hours_class_B,
    marks_class_B,
    color="red",
    marker="^",
    s=120,
    alpha=0.8,
    edgecolor="black",
    linewidth=1.5,
    label="Class B"
)

# Title
plt.title("Study Hours vs Marks", fontsize=16)

# Axis Labels
plt.xlabel("Study Hours", fontsize=12)
plt.ylabel("Marks", fontsize=12)

# Grid
plt.grid(True, linestyle="--", alpha=0.6)

# Legend
plt.legend(title="Classes")

# Axis Limits
plt.xlim(0, 8)
plt.ylim(30, 100)

# Show Plot
plt.show()
```
