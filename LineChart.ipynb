# Line Chart

## What is a Line Chart?

A **Line Chart** is a graph that represents data points connected by straight lines.

It is mainly used to visualize **changes or trends over an ordered sequence**, such as time.

Each point on the graph represents one observation, and the line connects these observations to make it easier to identify patterns.

---

## Why Do We Need a Line Chart?

Suppose a company records its monthly sales.

| Month | Sales |
|--------|-------|
| Jan | 120 |
| Feb | 150 |
| Mar | 180 |
| Apr | 160 |
| May | 210 |

Although the table contains all the information, it is difficult to identify the overall trend immediately.

A Line Chart transforms these numbers into a visual representation, making it easy to observe:

- Growth
- Decline
- Seasonal patterns
- Sudden increases
- Sudden decreases

---

## When Should We Use a Line Chart?

Use a Line Chart when:

- Data is ordered.
- Data changes over time.
- You want to visualize trends.
- You want to compare trends of multiple datasets.

---

## Real-Life Examples

- Stock Prices
- Temperature Changes
- Monthly Sales
- Website Visitors
- Rainfall
- Electricity Consumption
- Population Growth

---

## When NOT to Use a Line Chart?

Avoid using a Line Chart when:

- Data is categorical.
- Categories have no natural order.

Example:

❌ Product A, Product B, Product C

For categorical comparisons, a **Bar Chart** is more appropriate.

---

## Syntax

```python
plt.plot(x, y)
```

Where:

- `x` → Values for x-axis
- `y` → Values for y-axis

Matplotlib automatically joins the points using straight lines.

---

## Advantages

- Easy to identify trends.
- Easy to compare multiple datasets.
- Best for time-series data.
- Simple and easy to understand.

---

## Limitations

- Not suitable for categorical comparisons.
- Too many lines can make the graph cluttered.
- Does not show distribution of data.

````python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]

sales = [120, 150, 180, 160, 210]

plt.plot(months, sales)

plt.show()
````

````python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

sales = [120, 150, 180, 160, 210, 250]

plt.figure(figsize=(9,5))

plt.plot(

    months,
    sales,

    color="royalblue",

    linestyle="--",

    linewidth=3,

    marker="o",

    markersize=10,

    markerfacecolor="yellow",

    markeredgecolor="black",

    markeredgewidth=2,

    alpha=0.9,

    label="Monthly Sales"

)

plt.title(
    "Monthly Sales Analysis",
    fontsize=16,
    fontweight="bold",
    color="darkblue"
)

plt.xlabel(
    "Months",
    fontsize=12
)

plt.ylabel(
    "Sales (Units)",
    fontsize=12
)

plt.grid(
    True,
    linestyle=":",
    alpha=0.6
)

plt.legend(
    loc="upper left",
    fontsize=11
)

plt.xlim(-0.5,5.5)

plt.ylim(100,270)

plt.show()
````

# Multiple Line Chart

````python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun"]

product_A = [120,150,180,160,210,250]
product_B = [100,130,170,190,220,240]

plt.figure(figsize=(9,5))

plt.plot(
    months,
    product_A,
    marker="o",
    linewidth=3,
    label="Product A"
)

plt.plot(
    months,
    product_B,
    marker="s",
    linewidth=3,
    label="Product B"
)

plt.title("Sales Comparison")

plt.xlabel("Months")

plt.ylabel("Sales")

plt.grid(True)

plt.legend()

plt.show()
````
