# Saving Figures in Matplotlib

## What is `savefig()`?

The `savefig()` function is used to save a graph as an image or document file.

Instead of only displaying the graph on the screen, `savefig()` stores it permanently on your computer.

The saved figure can later be:

- Added to reports
- Uploaded to GitHub
- Used in presentations
- Shared on LinkedIn
- Included in dashboards
- Printed

---

## Why Do We Need `savefig()`?

Normally, when we use:

```python
plt.show()
```

the graph is only displayed on the screen.

After closing the graph, it is not saved automatically.

To keep the graph permanently, we use:

```python
plt.savefig()
```

---

## Syntax

```python
plt.savefig("filename.png")
```

The filename can include:

- File name
- Folder path
- File extension

Examples:

```python
plt.savefig("sales.png")

plt.savefig("Images/sales.png")

plt.savefig("Report.pdf")
```

---

## Supported File Formats

Matplotlib supports many formats, including:

- PNG
- JPG / JPEG
- PDF
- SVG
- EPS

The format is determined by the file extension.

````python
# savefig('filename.extension', dpi = value, bbox_inches = 'tights') 

import matplotlib.pyplot as plt



x = [1,2,3,4]
y = [10,20,15,25]

#create plot 
plt.plot(x,y,color='blue', marker= 'o')
plt.title('simple line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.savefig('line_plot.png', dpi= 300,bbox_inches ='tight' )

plt.show()
````

````python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]

sales = [120,150,180,170,220]

plt.figure(figsize=(8,5))

plt.plot(
    months,
    sales,
    color="royalblue",
    marker="o",
    linewidth=3,
    label="Sales"
)

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales")

plt.grid(True)

plt.legend()

plt.savefig(

    "monthly_sales.png",

    dpi=300,

    bbox_inches="tight",

    facecolor="white"

)

plt.show()
````

| Parameter             | Purpose                                         |
| --------------------- | ----------------------------------------------- |
| `dpi`                 | Resolution of the saved image                   |
| `bbox_inches="tight"` | Removes extra white space around the figure     |
| `facecolor`           | Background color of the saved figure            |
| `transparent=True`    | Saves with a transparent background             |
| `format`              | Specifies the file format explicitly (optional) |
