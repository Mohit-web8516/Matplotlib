# Scatter Plot

## What is a Scatter Plot?

A **Scatter Plot** is a graph that displays data as individual points on a two-dimensional coordinate system.

Each point represents one observation with:

- One value on the x-axis
- One value on the y-axis

Unlike a Line Chart, the points are **not connected** because every point represents an independent observation.

---

## Why Do We Need a Scatter Plot?

Suppose we collected data from students.

| Study Hours | Marks |
|-------------|------:|
| 1 | 35 |
| 2 | 45 |
| 3 | 55 |
| 4 | 65 |
| 5 | 72 |
| 6 | 80 |

Although the table contains all the information, it is difficult to understand the relationship between study hours and marks.

A Scatter Plot immediately helps us answer questions like:

- Do marks increase as study hours increase?
- Are there any unusual students?
- Is there any pattern?
- Is there any relationship?

---

## What Does a Scatter Plot Show?

A Scatter Plot is mainly used to study the **relationship (correlation)** between two numerical variables.

It helps us identify:

- Positive Relationship
- Negative Relationship
- No Relationship
- Outliers
- Clusters

---

## When Should We Use a Scatter Plot?

Use a Scatter Plot when:

- Both variables are numerical.
- You want to study relationships.
- You want to detect outliers.
- You want to identify clusters.
- You want to understand correlation.

---

## Real-Life Examples

- Height vs Weight
- Experience vs Salary
- Study Hours vs Marks
- Advertisement Cost vs Sales
- Temperature vs Ice Cream Sales
- House Size vs House Price

---

## When NOT to Use a Scatter Plot?

Avoid Scatter Plots when:

- Data is categorical.
- You only want to compare categories.
- You want to show percentages.
- You want to show trends over time.

---

## Syntax

```python
plt.scatter(x, y)
```

Where:

- `x` → Numerical values
- `y` → Numerical values

Each `(x, y)` pair becomes one point.

---

## Advantages

- Shows relationships clearly.
- Detects outliers.
- Detects clusters.
- Helps identify correlations.
- Useful before Machine Learning.

---

## Limitations

- Does not show percentages.
- Not suitable for categorical data.
- Too many points can overlap.
- Correlation does NOT imply causation.

# Types of Correlation

---

## 2. Negative Correlation

# Basic Scatter Plot

````python
import matplotlib.pyplot as plt

study_hours = [1,2,3,4,5,6]

marks = [35,45,55,65,72,80]

plt.scatter(
    study_hours,
    marks
)

plt.show()
````

# Professional Scatter Plot

````python
import matplotlib.pyplot as plt

study_hours = [1,2,3,4,5,6,7,8]

marks = [35,45,55,65,72,80,88,95]

plt.figure(figsize=(8,5))

plt.scatter(

    study_hours,

    marks,

    color="royalblue",

    marker="o",

    s=150,

    alpha=0.8,

    edgecolor="black",

    linewidth=2,

    label="Students"

)

plt.title(
    "Study Hours vs Marks",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Study Hours",
    fontsize=12
)

plt.ylabel(
    "Marks",
    fontsize=12
)

plt.grid(
    True,
    linestyle="--",
    alpha=0.5
)

plt.legend()

plt.xlim(0,9)

plt.ylim(20,100)

plt.show()
````

# Scatter Plot with Two Datasets

````python
import matplotlib.pyplot as plt

hours_A = [1,2,3,4,5,6]

marks_A = [35,45,55,65,72,80]

hours_B = [1,2,3,4,5,6]

marks_B = [30,42,50,60,67,75]

plt.figure(figsize=(8,5))

plt.scatter(
    hours_A,
    marks_A,

    color="blue",

    marker="o",

    s=120,

    label="Class A"
)

plt.scatter(
    hours_B,
    marks_B,

    color="red",

    marker="^",

    s=120,

    label="Class B"
)

plt.title("Study Hours vs Marks")

plt.xlabel("Study Hours")

plt.ylabel("Marks")

plt.grid(True)

plt.legend()

plt.show()
````

# Interview Questions
# Q1 What is a Scatter Plot?

A Scatter Plot displays the relationship between two numerical variables using individual points.

# Q2 When should you use a Scatter Plot?

When studying relationships, correlations, clusters, or outliers between two numerical variables.

# Q3 What is correlation?

Correlation describes how two variables change with respect to each other.

Positive
Negative
No Correlation
# Q4 Does a Scatter Plot prove causation?

No.

A Scatter Plot only shows a relationship (correlation) between variables. It does not prove that one variable causes the other to change.

# Q5 Why is Scatter Plot important in Machine Learning?

Because it helps us:

Understand relationships between variables.
Detect outliers.
Explore data before building predictive models.

# 📌 One important concept

A common misconception is that a scatter plot shows a trend over time. It doesn't. Its primary purpose is to show the relationship between two numerical variables. If your x-axis represents time or another naturally ordered sequence and you want to emphasize how values change over that sequence, a line chart is usually the better choice.
